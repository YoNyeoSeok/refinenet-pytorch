```python
from torchvision import transforms
import numpy as np
transform = transforms.Compose([transforms.ToTensor(),
                                transforms.Lambda(lambda img: img*255)])
foggy_cityscapes = EnvCityscapes('/workbench/data/Cityscapes', env=["", 'foggy', 'foggyDBF', 'transmittanceDBF'], env_type=[[''], ['beta_0.02', 'beta_0.01'], ['beta_0.02'], ['beta_0.02']], target_type=['semantic', 'color'], split='val',
                                transform=transforms.Lambda(lambda imgs: tuple(map(transform, imgs))),
                                target_transform=transforms.Lambda(lambda imgs: tuple(map(np.array, imgs))))
```

```python
from matplotlib import pyplot as plt

for i, (x, y) in enumerate(foggy_cityscapes):
    print(i, len(x), len(y))
    for x_ in x:
        print(x_.shape)
        plt.imshow(x_.byte().numpy().transpose(1, 2, 0))
        plt.show()
    for x_ in y:
        print(x_.shape)
        plt.imshow(x_)
        plt.show()
    break
```
